```python
#Her er menneskene helt trygge, så nå er det bare dødsfall via naturlig årsaker og fødsel og innvandring som endrer populasjonen
from pylab import *

fm = 500       #Dette er friske mennesker som ikke har blitt smittet. Vi antar at alle er mellom 20 og 40 år, siden det bare er plass til 500. Disse vil være i best form og har høyest sannslighet til å overleve vanlige sykdommer.
z = 0          #Dette er antall smittede zombier, som ikke er noen ennå
fr = 1.36E-4   #Barn født per kvinne, i apokalypsen. En kvinne får gjennomsnittlig 2 barn i løpet av livet. Kvinnene her er mellom 20 og 40 år, når de fleste får barn. Vi delte 2 på 20(for de 20 årene mellom), så tok svaret og delte på 365 dager og delte det på 2 igjen for ca halvparten er kvinner
nb = 3         #med tanke på at det bor rundt 150 000 mennesker i Alexandria og mange rundt, vil denne være veldig høy ved start. Ettersom flere vil bli smittet, vil denne verdien avta
dr = 5E-5    #Dødsraten for å dø for en person mellom 20 og 40 år. Dette er litt ca fra statistisk sentralbyrå

pop = [500]
dager = [0]
for i in range(1,365):
    fm = (fm * fr + nb) -dr
    pop.append(fm)
    dager.append(i)

print("Populasjon: ", pop)
print("Antall dager: ", dager)
plot(dager, pop)
```

```python
from pylab import *

m = 500                     #Mennesker
fr = 1 + ((m/70)/12)/m      #fødselsraten, tidskalaen er målt i måneder, og man regner med menneskene lever i 70 år
dr = fr*0.95                #dødsraten
innv = 3                     #innvandring i måneden
tid = 0
maxm = 10000

#lister
antall_mnd = []
populasjon_liste = []

def population(n,m):        #n står for number of months
    global fr, dr, innv
    m0=m
    fr = fr*(1-(m0/maxm**2))
    innv = innv*(1-(m0/maxm**2))
    for i in range(n):
        m0 += m0*(fr-dr) + innv
    return m0

#print(population(60,m))

for i in range(100):
    antall_mnd.append(i)
    populasjon_liste.append(population(i,m))
    
plot(antall_mnd, populasjon_liste)


```
