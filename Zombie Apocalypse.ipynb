```python
#Her er menneskene helt trygge, s√• n√• er det bare d√∏dsfall via naturlig √•rsaker og f√∏dsel og innvandring som endrer populasjonen
from pylab import *

fm = 500       #Dette er friske mennesker som ikke har blitt smittet. Vi antar at alle er mellom 20 og 40 √•r, siden det bare er plass til 500. Disse vil v√¶re i best form og har h√∏yest sannslighet til √• overleve vanlige sykdommer.
z = 0          #Dette er antall smittede zombier, som ikke er noen enn√•
fr = 1.36E-4   #Barn f√∏dt per kvinne, i apokalypsen. En kvinne f√•r gjennomsnittlig 2 barn i l√∏pet av livet. Kvinnene her er mellom 20 og 40 √•r, n√•r de fleste f√•r barn. Vi delte 2 p√• 20(for de 20 √•rene mellom), s√• tok svaret og delte p√• 365 dager og delte det p√• 2 igjen for ca halvparten er kvinner
nb = 3         #med tanke p√• at det bor rundt 150 000 mennesker i Alexandria og mange rundt, vil denne v√¶re veldig h√∏y ved start. Ettersom flere vil bli smittet, vil denne verdien avta
dr = 5E-5    #D√∏dsraten for √• d√∏ for en person mellom 20 og 40 √•r. Dette er litt ca fra statistisk sentralbyr√•

pop = [500]
dager = [0]
for i in range(1,365):
    fm = (fm * fr + nb) -dr
    pop.append(fm)
    dager.append(i)

print("Populasjon: ", pop)
print("Antall dager: ", dager)
plot(dager, pop)
```

```python
from pylab import *

m = 500                     #Mennesker
fr = 1 + ((m/70)/12)/m      #f√∏dselsraten, tidskalaen er m√•lt i m√•neder, og man regner med menneskene lever i 70 √•r
dr = fr*0.95                #d√∏dsraten
innv = 3                     #innvandring i m√•neden
tid = 0
maxm = 10000

#lister
antall_mnd = []
populasjon_liste = []

def population(n,m):        #n st√•r for number of months
    global fr, dr, innv
    m0=m
    fr = fr*(1-(m0/maxm**2))
    innv = innv*(1-(m0/maxm**2))
    for i in range(n):
        m0 += m0*(fr-dr) + innv
    return m0

#print(population(60,m))

for i in range(100):
    antall_mnd.append(i)
    populasjon_liste.append(population(i,m))
    
plot(antall_mnd, populasjon_liste)


```
